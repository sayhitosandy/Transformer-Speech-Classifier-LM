```python
import json
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def get_csv_from_json(file_name):
    with open(f"{file_name}.json", "r") as f:
        data = json.load(f)

    df = pd.DataFrame(data["history"]).round(4)
    df.to_csv(f"{file_name}.csv", index=False)

    return df
```

```python
def plot_cls(df, title, file_path):
    plt.plot(df["epoch"], df["train_accuracy"], label="Train", marker='.')
    plt.plot(df["epoch"], df["test_accuracy"], label="Test", marker='.')
    plt.grid()
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.savefig(file_path, dpi=360)
    plt.show()
```

```python
def plot_compare_cls(df1, df2, comparison, title, file_path):
    plt.plot(df1["epoch"], df1["train_accuracy"], label=f"Train", marker='.')
    plt.plot(df2["epoch"], df2["train_accuracy"], label=f"Train ({comparison})", marker='*')
    plt.plot(df1["epoch"], df1["test_accuracy"], label=f"Test", marker='.')
    plt.plot(df2["epoch"], df2["test_accuracy"], label=f"Test ({comparison})", marker='*')
    plt.grid()
    plt.xlabel("Epochs")
    plt.ylabel("Accuracy")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.savefig(file_path, dpi=360)
    plt.show()    
```

```python
def plot_lm(df, title, file_path):
    plt.plot(df["epoch"], df["train_perplexity"], label="Train", marker='.')
    plt.plot(df["epoch"], df["hbush_test_perplexity"], label="HBush Test", marker='.')
    plt.plot(df["epoch"], df["obama_test_perplexity"], label="Obama Test", marker='.')
    plt.plot(df["epoch"], df["wbush_test_perplexity"], label="WBush Test", marker='.')
    plt.grid()
    plt.xlabel("Epochs")
    plt.ylabel("Perplexity")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.savefig(file_path, dpi=360)
    plt.show()
```

```python
def plot_compare_lm(df1, df2, comparison, title, file_path):
    plt.plot(df1["epoch"], df1["train_perplexity"], label="Train", marker='.')
    plt.plot(df2["epoch"], df2["train_perplexity"], label=f"Train ({comparison})", marker='*')

    plt.plot(df1["epoch"], df1["hbush_test_perplexity"], label="HBush Test", marker='.')
    plt.plot(df2["epoch"], df2["hbush_test_perplexity"], label=f"HBush Test ({comparison})", marker='*')

    plt.plot(df1["epoch"], df1["obama_test_perplexity"], label="Obama Test", marker='.')
    plt.plot(df2["epoch"], df2["obama_test_perplexity"], label=f"Obama Test ({comparison})", marker='*')

    plt.plot(df1["epoch"], df1["wbush_test_perplexity"], label="WBush Test", marker='.')
    plt.plot(df2["epoch"], df2["wbush_test_perplexity"], label=f"WBush Test ({comparison})", marker='*')

    plt.grid()
    plt.xlabel("Epochs")
    plt.ylabel("Perplexity")
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.savefig(file_path, dpi=360)
    plt.show()
```
